```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from pantherapy.panthera import CrossSection

y = np.array([90, 80, 78, 70, 71, 79, 81, 91])
z = np.array([110, 120, 200, 210, 230, 240, 350, 360])
roughness = 0.013
xs = CrossSection(y, z, roughness)

fig, ax = plt.subplots()

fig.set_size_inches((11, 8.5))

_ = xs.plot(80.5, ax)
```
