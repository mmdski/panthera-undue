```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from pantherapy import CrossSection
from pantherapy.plot import CrossSectionPlot
```

```python
y = np.array([90, 80, 78, 70, 71, 79, 81, 91])
z = np.array([110, 120, 200, 210, 230, 240, 350, 360])
roughness = np.array([0.013])
xs = CrossSection(y, z, roughness)
xs_plot = CrossSectionPlot(xs)
```

```python
fig, ax = plt.subplots()

fig.set_size_inches((11, 8.5))

_ = xs_plot.plot(80.5, ax)
```
