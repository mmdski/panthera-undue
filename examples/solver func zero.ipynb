```python
%matplotlib inline

from pantherapy.panthera import CrossSection


y_xs = [3, 0, 0, 3]
z_xs = [0, 4.5, 8, 12.5]
roughness = 0.015

depth = 1

xs = CrossSection(y_xs, z_xs, roughness)
_ = xs.plot(depth)
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from pantherapy.reach import Reach


slope = 0.001

dx = 12.5

x0 = dx
y0 = 0

x1 = 0
y1 = dx*slope

reach = Reach()
reach.put(xs, x0, y0)
reach.put(xs, x1, y1)

flow = 4

H0 = 1
h0 = H0 + y0

H1 = np.logspace(-5, np.log10(2), 200)
h1 = H1 + y1

f = reach.energy_diff(h1, flow, 0, h0, flow, 1) - y1

ax = plt.axes()

_ = ax.plot(H1, f)
_ = ax.set_xlim(0, 2)
_ = ax.set_ylim(-4.0, 3.5)
_ = ax.axhline(c='k')
_ = ax.set_xlabel('$H_1$, in meters')
_ = ax.set_ylabel('$f\\left(H_1\\right)$')
```
