```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from pantherapy import CrossSection

y = np.array([90, 80, 78, 70, 71, 79, 81, 91])
z = np.array([110, 120, 200, 210, 230, 240, 350, 360])
roughness = np.array([0.013])

_ = plt.plot(z, y)
_ = plt.xlabel('z')
_ = plt.ylabel('y')
```

```python
xs = CrossSection(y, z, roughness)
wse = np.linspace(y.min(), y.max())
```

```python
area = xs.area(wse)

fig, ax = plt.subplots()
fig.set_size_inches((11, 8.5))

_ = ax.plot(wse, area, label='area')
_ = ax.set_xlabel('wse')
_ = ax.set_ylabel('area')
```

```python
top_width = xs.top_width(wse)
wetted_perimeter = xs.wetted_perimeter(wse)

fig, ax = plt.subplots()
fig.set_size_inches((11, 8.5))

_ = ax.plot(wse, top_width, label='top width')
_ = ax.plot(wse, wetted_perimeter, label='wetted perimeter')
_ = ax.legend()
_ = ax.set_xlabel('wse')
```

```python
hydraulic_depth = xs.hydraulic_depth(wse)
hydraulic_radius = xs.hydraulic_radius(wse)

fig, ax = plt.subplots()
fig.set_size_inches((11, 8.5))

_ = ax.plot(wse, hydraulic_depth, label='hydraulic depth')
_ = ax.plot(wse, hydraulic_radius, label='hydraulic radius')
_ = ax.legend()
_ = ax.set_xlabel('wse')
```
