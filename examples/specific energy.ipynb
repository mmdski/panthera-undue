```python
import numpy as np

from pantherapy.panthera import CrossSection

y = np.array([5, 0, 0, 0, 5])
z = np.array([0, 0, 0.5, 1, 1])
roughness = 0.030
xs = CrossSection(y, z, roughness)
```

```python
y = np.linspace(0.25, 3.6, 100)
q = 3
specific_energy = xs.specific_energy(y, q)
yc = xs.critical_depth(q)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

_ = plt.plot(specific_energy, y)
_ = plt.axhline(yc, label='Critical depth', ls='--', color='k')
_ = plt.axvline(specific_energy.min(), ls=':', label='Minimum E', color='k')
_ = plt.xlabel('Specific energy')
_ = plt.ylabel('Depth')
_ = plt.xlim((1, 3.5))
_ = plt.legend()
```
